```python
#1.How many people have heart disesase and how many people doesn't have haert disesase?

#getting the values
df.target.value_counts()
```

```python
#plotting bar chart
df.target.value_counts().plot(kind='bar',color=["red","green"])
plt.title("Heart disesase values")
plt.xlabel("1=Heart Disease,0=No Heart Disease")
plt.ylabel("Amount")
plt.show()
```

```python
#getting the values
#In sex column part '0' means female and '1' means male
#In Target column part '0' means no heart disesase '1' means heart disesase

df.sex.value_counts()
```

```python
#plotting pie chart
df.sex.value_counts().plot(kind='pie',figsize=(8,6))
plt.title("Male Female Ratio")
plt.legend(['Male','Female'])
plt.show()
```

```python
#2.People of which sex has most heart disease?

pd.crosstab(df.target,df.sex)
```

```python
sns.countplot(x='target', data=df,hue='sex')
plt.title("Heart Disease Frequency for Sex")
plt.xlabel("0=No heart disesase,1=Heart disease")
plt.show()
```

```python
#3.People of which sex has which type of chest pain most?

#counting values for different chest pain
df.cp.value_counts()
```

```python
#plotting a bar chart
df.cp.value_counts().plot(kind='bar',color=['salmon', "lightblue","yellow","purple"])
plt.title("chest pain vs count")
plt.xlabel("chest pain type")
plt.show()
```

```python
pd.crosstab(df.sex,df.cp)
```

```python
pd.crosstab(df.sex,df.cp).plot(kind='bar',color=["coral","silver","plum","grey"])
plt.title("type of chest pain foe sex")
plt.xlabel("0=Female,1=Male")
plt.show()
```

```python
#5.What is the average cholesterol level for people with and without heart disesase?

df.chol.value_counts()
```
