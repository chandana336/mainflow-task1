```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
import os 
os.chdir("F:\\")
data=pd.read_csv("householdtask3.csv")
```

```python
data.head(10)
```

```python
#scatterplot with year against own
plt.scatter(data['year'],data['own'])
#adding title to the plot
plt.title("scatterplot")
# setting the x abd y labels
plt.xlabel('year')
plt.ylabel('own')
#showing result
plt.show()
```

```python
#line chart with year and own
plt.plot(data['year'])
plt.plot(data['own'])
# adding title to plot
plt.title("line chart")
#setting x and y labels
plt.xlabel('year')
plt.ylabel('own')
#showing result
plt.show()
```

```python
#bar chart or barplot
plt.bar(data['year'],data['own'])
#adding title to the plot
plt.title("barchart")
#setting x and y labels
plt.xlabel('year')
plt.ylabel('own')
#showing result
plt.show()
```

```python
#histogram
plt.hist(data['income'])
plt.title("histogram")
plt.show()
```
